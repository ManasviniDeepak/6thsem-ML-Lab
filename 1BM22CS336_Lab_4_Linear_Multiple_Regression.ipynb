# linear_regression_housing_area_price

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('/content/drive/MyDrive/housing_area_price.csv')

# Quick preview of the data
print(df.head())

# Plot the data points
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (Rs)')
plt.scatter(df.area, df.price, color='red', marker='+')

# Prepare the data for training
X = df[['area']]  # Features (must be a 2D array)
y = df.price      # Target

# Create and train the linear regression model
reg = LinearRegression()
reg.fit(X, y)

# Coefficient (m) and intercept (b)
print(f"Coefficient (m): {reg.coef_[0]}")
print(f"Intercept (b): {reg.intercept_}")

# Predict price for an area of 3300 sq ft
area_3300_pred = reg.predict([[3300]])
print(f"Predicted price for 3300 sq ft: {area_3300_pred[0]}")

# Manual calculation for confirmation
manual_calc = 3300 * reg.coef_[0] + reg.intercept_
print(f"Manual calculation: {manual_calc}")

# Predict price for an area of 5000 sq ft
area_5000_pred = reg.predict([[5000]])
print(f"Predicted price for 5000 sq ft: {area_5000_pred[0]}")

plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(X), color='blue')  # This draws the line
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (Rs)')
plt.show()
```

# Cannada per capita Income

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
df = pd.read_csv('/content/drive/MyDrive/canada_per_capita_income.csv')
df.head()
df.rename(columns={'per capita income (US$)': 'income'}, inplace=True)
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(df.year, df.income, color='red', marker='+')
X = df[['year']]  # Features (2D array)
y = df['income']  # Target (1D array)
model = LinearRegression()
model.fit(X, y)
year_2020_income = model.predict([[2020]])
print(f"Predicted per capita income in 2020: {year_2020_income[0]:.2f} US$")
print(f"Coefficient (m): {model.coef_[0]}")
print(f"Intercept (b): {model.intercept_}")
```

# Salary of Employee

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/salary.csv')

# Clean/rename columns if necessary
df.rename(columns={'YearsExperience': 'experience', 'Salary': 'salary'}, inplace=True)
df.experience=df.experience.fillna(df.experience.median())
df.salary=df.salary.fillna(df.salary.median())
# Plot data
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.scatter(df.experience, df.salary, color='red', marker='+')

# Prepare data
X = df[['experience']]
y = df['salary']

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Predict salary for 12 years of experience
predicted_salary = model.predict([[12]])
print(f"Predicted salary for 12 years of experience: {predicted_salary[0]:.2f} USD")

# Optional: Show coefficient and intercept
print(f"Coefficient (m): {model.coef_[0]}")
print(f"Intercept (b): {model.intercept_}")

# Plot regression line
plt.plot(df.experience, model.predict(X), color='blue')
plt.show()
```

# MULTIPLE LINEAR REGRESSION: hiring.csv

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/hiring.csv')

# Rename columns for easier access
df.rename(columns={
    'test_score(out of 10)': 'test_score',
    'interview_score(out of 10)': 'interview_score',
    'salary($)': 'salary'
}, inplace=True)

# Convert experience column from text to numbers
def convert_to_number(word):
    word_to_number = {
        'zero': 0,
        'one': 1,
        'two': 2,
        'three': 3,
        'four': 4,
        'five': 5,
        'six': 6,
        'seven': 7,
        'eight': 8,
        'nine': 9,
        'ten': 10,
        'eleven': 11,
        'twelve': 12
    }
    return word_to_number.get(str(word).lower(), 0)

df['experience'] = df['experience'].apply(convert_to_number)

# Handle missing values
df.test_score = df.test_score.fillna(df.test_score.median())
df.experience = df.experience.bfill()

# Prepare data
X = df[['experience', 'test_score', 'interview_score']]
y = df['salary']

# Train model
model = LinearRegression()
model.fit(X, y)
print("\nCoefficient: ",model.coef_)
print("\nIntercept: ",model.intercept_)
# Predict salaries
salary_1 = model.predict([[2, 9, 6]])
salary_2 = model.predict([[12, 10, 10]])

print(f"\nPredicted salary for 2 yr exp, 9 test score, 6 interview score: {salary_1}")
print(f"Predicted salary for 12 yr exp, 10 test score, 10 interview score: {salary_2}")
```

# MULTIPLE LINEAR REGRESSION: 100_companies.csv

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/1000_Companies.csv')

# Prepare features and target
X = df[['R&D Spend', 'Administration', 'Marketing Spend', 'State']]
y = df['Profit']

# One-hot encode 'State'
ct = ColumnTransformer([('encoder', OneHotEncoder(), ['State'])], remainder='passthrough')
X_encoded = ct.fit_transform(X)

# Train model
model = LinearRegression()
model.fit(X_encoded, y)

# Predict profit for given input:
# Florida -> [0, 1, 0]
# R&D Spend = 91694.48
# Administration = 515841.3
# Marketing Spend = 11931.24
input_data = [[0, 1, 0, 91694.48, 515841.3, 11931.24]]

predicted_profit = model.predict(input_data)
print("Coefficient: ",model.coef_)
print("Intercept: ",model.intercept_)
print(f"Predicted profit: {predicted_profit[0]:.2f} USD")
```
